```python
import pandas as pd
```

```python
df = pd.read_csv('test-groupby.csv')
```

```python
df
```

```python
col = 'grade'
```

```python
df.groupby([col])
```

```python
for name, group in df.groupby([col]):
    print(name)
    display(group.head())
```

```python
df['mean'] = df.groupby([col])['target'].transform(lambda x: x.mean())
df
```

```python
df.groupby([col])['target'].agg(['mean'])
```

```python
df.groupby([col])['target'].agg(['mean']).reset_index() #groupby对象 变成df
```

```python
temp = df.groupby([col])['target'].agg(['mean']).reset_index().rename(columns={'mean': col + '_target_mean'})
temp
```

```python
temp.index = temp[col].values
temp.index
```

```python
temp = temp[col + '_target_mean'].to_dict()
temp
```

```python
df[col + '_target_mean'] = df[col].map(temp)
df
```

```python
item = 'n4'
```

```python
df['grade'] = df['grade'].map({'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7})
```

```python
df
```

```python
for name, group in df.groupby([item]):
    print(name)
    display(group)
```

```python
df.groupby([item])['grade'].transform('mean')
```

```python
df['grade_to_mean_' + item] = df['grade'] / df.groupby([item])['grade'].transform('mean')
```
